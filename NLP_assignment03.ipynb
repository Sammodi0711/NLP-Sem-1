```python
import nltk
from nltk.corpus import brown
from nltk.tokenize import word_tokenize

nltk.download('brown')
nltk.download('punkt')
nltk.download('punkt_tab')

romanceword = brown.words(categories='romance')
romancesentence=''.join(romanceword)
print(romancesentence)

token=word_tokenize(romancesentence)
print(token)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.probability import FreqDist

frequencyDistribution = FreqDist()

for word in word_tokenize(romancesentence):
  if word.isalpha():
     frequencyDistribution[word.lower()] += 1

top20Words = frequencyDistribution.most_common(20)
print(top20Words)

# Bar Chart
words, counts = zip(*top20Words)
plt.bar(words, counts, color='grey')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 20 Words - Romance Category')
plt.xticks(rotation=45)
plt.show()

print()

# Word Cloud
wc = WordCloud(width=800, height=800, background_color='white').generate_from_frequencies(dict(top20Words))
plt.figure(figsize=(5, 8))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Q3.Apply Stemming (PorterStemmer)
#Task: Apply NLTK’s PorterStemmer to the tokenized words, then list each stemmed word along with its frequency.

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmedwords = []

for word in token:
  if word.isalpha():
    stemmedwords.append(stemmer.stem(word.lower()))

frequencyDistributionstems = FreqDist(stemmedwords)

for word,freq in frequencyDistributionstems.items():
  print(word,freq)
```

```python
# Q4. Apply Lemmatization (WordNetLemmatizer)
#Task: Apply NLTK’s WordNetLemmatizer to the tokenized words, then list each lemmatized word along with its frequency.

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmatizedwords = []
for word in token:
   lemmatizedwords.append(lemmatizer.lemmatize(word))

FriquencyDistributionLemma = FreqDist(lemmatizedwords)

for word,freq in FriquencyDistributionLemma.items():
  print(word,freq)
```

```python
# Q5. Compare Stemming vs Lemmatization
#Task: Create a side-by-side table or bar chart showing how stemming and lemmatization have reduced words differently (compare counts for at least 15 common words)

from collections import Counter

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Sample text
text = """
The boys were playing in the park while the girls were reading books.
Running and jumping made them very happy.
Cats are chasing the mice, and the dogs are barking loudly.
The teacher was explaining the lessons to the students.
Better tools make the work easier and faster.
"""
words = nltk.word_tokenize(text.lower())

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_words = [stemmer.stem(word) for word in words if word.isalpha()]
lemmatized_words = [lemmatizer.lemmatize(word) for word in words if word.isalpha()]

stem_counts = Counter(stemmed_words)
lemma_counts = Counter(lemmatized_words)

common_words = [w for w, _ in Counter(words).most_common() if w.isalpha()][:15]

stem_values = [stem_counts.get(stemmer.stem(w), 0) for w in common_words]
lemma_values = [lemma_counts.get(lemmatizer.lemmatize(w), 0) for w in common_words]

# Plot side-by-side bar chart
x = range(len(common_words))
plt.figure(figsize=(5,3))
plt.bar([i - 0.2 for i in x], stem_values, width=0.4, label='Stemming', color='orange')
plt.bar([i + 0.2 for i in x], lemma_values, width=0.4, label='Lemmatization', color='skyblue')
plt.xticks(x, common_words, rotation=45)
plt.xlabel('Original Words')
plt.ylabel('Frequency after Processing')
plt.title('Comparison: Stemming vs Lemmatization')
plt.legend()
plt.tight_layout()
plt.show()
```
